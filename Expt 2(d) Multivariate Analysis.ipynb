# Multivariate analysis of dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("archive/city_hour.csv")
df=df[["PM2.5","PM10","NO2","SO2","O3","AQI","AQI_Bucket"]]
df.head(20)
```

```python
sns.scatterplot(x='SO2',y='O3',hue='AQI',data=df,)
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.show()
```

```python
sns.heatmap(df.corr(method="pearson"),annot=True)
```

```python
sns.pairplot(df,hue='AQI')
```
